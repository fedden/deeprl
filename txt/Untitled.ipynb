```python
import pandas as pd
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

def get_info(filename):
    file = open(filename, "r") 
    lines = file.readlines() 
    ep_rewards = []
    fe_dists = []
    pa_dists = []

    i = 0
    for line in lines:
        if "Average Patch" in line:
            av_pa = float(line[26:].strip())
        elif "Average Feature" in line:
            av_fe = float(line[26:].strip())
        elif "Reward" in line:
            ep_rewards.append(float(line[20:].strip()))
        elif "Feature Distance" in line:
            fe_dists.append(float(line[20:].strip()))
        elif "Parameter Distance" in line:
            pa_dists.append(float(line[20:].strip()))

    return av_pa, av_fe, ep_rewards, fe_dists, pa_dists

or_av_pa, or_av_fe, or_ep_re, or_fe_di, or_pa_di = get_info("comparison.txt")
ne_av_pa, ne_av_fe, ne_ep_re, ne_fe_di, ne_pa_di = get_info("little_diff.txt")
```

```python
def display_hists(all_patch_distances, all_feature_distances, title):
    
    df = pd.DataFrame({'Patch Distances': np.array(all_patch_distances),
                       'Feature Distances': np.array(all_feature_distances)})
    df.hist(layout=(1, 2), grid=False, figsize=(8,4))
    plt.suptitle(title)
    plt.xlabel("Amount Absolute Error")
    plt.ylabel("Amount Examples")
    plt.show()
    
display_hists(or_pa_di, or_fe_di, "Start")
display_hists(ne_pa_di, ne_fe_di, "End")
```
